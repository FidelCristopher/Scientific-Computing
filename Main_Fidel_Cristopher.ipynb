```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
```

```python
x = [0, 1 , 2]
y = [1, 3, 2]

f = interp1d(x, y)

y_hat = f(1.5)
print(y_hat)
```

```python
plt.figure(figsize=(10, 8)) #Ukuran plow 10x8 inci
plt.plot(x, y, '-ob') #Menampilkan titik data dengan garis biru dan lingkaran
plt.plot(1.5, y_hat, 'ro') #Menampilkan titik hasil interpolasi dengan warna merah
plt.title('Linear Interpolation at x = 1.5') #Judul grafik
plt.xlabel('x') #Label sumbu x
plt.ylabel('y') #Label sumbu y
plt.show() #Menampilkan grafik
```

```python
import numpy as np
import matplotlib.pyplot

def divided_diff(x, y):
  n = len(y)
  coef = np.zeros([n, n]) #Membuat tabel kosong untuk divided differences
  coef[:, 0] = y #Kolom pertama berisi nilai y

  for j in range (1, n): #Loop untuk mengisi kolom-kolom berikutnya
    for i in range(n - j):
      coef[i][j] = (coef[i + 1][j - 1] - coef[i][j - 1]) / (x[i + j] - x[i])

  return coef

def newton_poly(coef, x_data, x):
  n = len(x_data) - 1 #Derajat polinomial
  p = coef[n]

  for k in range(1, n + 1):
    p = coef[n - k] + (x - x_data[n - k]) * p

  return p
```

```python
x = np.array([-5, -1, 0, 2])
y = np.array([-2, 6, 1, 3])

#Menghitung koefisien divided difference
a_s = divided_diff(x, y)[0, :]

#Evaluasi interpolasi pada titik-titik baru
x_new = np.arange(-5, 2.1, .1)
y_new = newton_poly(a_s, x, x_new)

#Menampilkan plot
plt.figure(figsize=(12, 8))
plt.plot(x, y, 'bo') #Menampilkan titik data asli
plt.plot(x_new, y_new, 'b') #Menampilkan kurva interpolasi
```

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

x = [0, 1, 2] #Titik data pada sumbu x
y = [1, 3, 2] #Titik data pada sumbu y

#Membuat fungsi interpolasi kubuk dengan batas 'natrual'
f = CubicSpline(x, y, bc_type='natural')

#Membuat rentang nilai x untuk plot hasil interpolasi
x_new = np.linspace(0, 2, 100) #100 titik antara 0 dan 2
y_new = f(x_new) #Menghitung nilai y menggunakan interpolasi kubik

#Membuat plot
plt.figure(figsize=(10, 8)) #Ukuran plot 10x8 inci
plt.plot(x_new, y_new, 'b') #Menampilkan kurva interpolasi kubik dalam warna biru
plt.plot(x, y, 'ro') #Menampilkan titik data asli dalam warna merah
plt.title('Cubic Spline Interpolation') #Judul grafik
plt.xlabel('x') #Label sumbu x
plt.ylabel('y') #Label sumbu y
plt.show() #Menampilkan grafik
```

Olimpiade tahun 1990, juara lomba sprint 100m adalah Ben Johnson. Catatan larinya ketika 0 detik kecepatannya adalah 10m/detik. Pada saat 5 meter kecepatannya adalah 15m/detik. Pada saat 10m kecepatannya adalah 20m/detik. Berapa kecepatan larinya pada saat jarak 7.2m. Jawablah menggunakan coding python linear interpolation dibawah codingan sesi hari ini!

```python
from scipy.interpolate import interp1d
import numpy as np

# Modul lab

# Data jarak lari
x = np.array([0, 5, 10])
y = np.array([10, 15, 20])

f = interp1d(x, y, kind='linear')

# Estimate speed at 7.2 meters
speed_at_7_2m = f(7.2)

y_hat = f(7.2)
print(f"Kecepatan lari Ben Johnson pada jarak 7.2 meter: {speed_at_7_2m} m/detik")
```

```python
plt.figure(figsize=(10, 8)) # Ukuran plot 10x8 inci
plt.plot(x, y, '-ob') # Menampilkan titik data dengan garis biru dan lingkaran
plt.plot(7.2, y_hat, 'ro') # Menampilkan titik haris interpolasi dengan garis merah dan lingkaran
plt.title('Lienar interpolation at x = 7.2') # Judul grarik
plt.xlabel('x') # Nama label sumbu x
plt.ylabel('y') # Nama label sumbu y
plt.show() # Menampilkan grafik
```
